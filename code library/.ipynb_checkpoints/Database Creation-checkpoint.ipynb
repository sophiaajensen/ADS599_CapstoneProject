```python
#Create a database containing only the questions that were consistently asked every year; used chatgpt on 7/20/2024 to debug code and filter to another database

# Connect to database
db_path = "/Users/duynguyen/Documents/Documents - DUY’s Mac mini/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"
conn = sqlite3.connect(db_path)

# Step 1: Find question_IDs that appear in all 7 years
df_consistent_qids = pd.read_sql("""
    SELECT question_ID
    FROM survey_question_master
    GROUP BY question_ID
    HAVING COUNT(DISTINCT year) = 7
""", conn)

# Step 2: Filter original table using these consistent question_IDs
df_filtered = pd.read_sql("""
    SELECT *
    FROM survey_question_master
    WHERE question_ID IN (
        SELECT question_ID
        FROM survey_question_master
        GROUP BY question_ID
        HAVING COUNT(DISTINCT year) = 7
    )
""", conn)

# Step 3: Save result to a new table
df_filtered.to_sql("survey_question_consistent_18_24", conn, if_exists="replace", index=False)

# Step 4: Verify row count
df_check = pd.read_sql("SELECT COUNT(*) as total_rows FROM survey_question_consistent_18_24", conn)
print(df_check)

# Close connection
conn.close()
```

```python
# Paths to DBs
source_db = "/Users/duynguyen/Documents/Documents - DUY’s Mac mini/GitHub/ADS599_CapstoneProject/data/fevs_data.db"
cleaned_db = "/Users/duynguyen/Documents/Documents - DUY’s Mac mini/GitHub/ADS599_CapstoneProject/data/fevs_data_cleaned.db"

# Use your DataFrame of consistent question IDs
question_ids = df_consistent_question_id['question_ID'].dropna().astype(str).tolist()

# Connect to databases
conn_src = sqlite3.connect(source_db)
conn_dst = sqlite3.connect(cleaned_db)

# Process for each year
for year in range(2018, 2025):
    table_name = f"fevs_{year}"
    new_table_name = f"fevs_cleaned_{year}"

    # Get all columns from the original table
    df_info = pd.read_sql(f"PRAGMA table_info({table_name});", conn_src)
    all_columns = df_info['name'].tolist()

    # Exclude columns named 'agency' or 'AGENCY' (case-insensitive)
    all_columns = [col for col in all_columns if col.lower() != 'agency']

    # Keep only columns that do not start with Q, V, AN, an
    metadata_columns = [col for col in all_columns if not col.startswith("Q") and not col.startswith("V") and not col.startswith("AN") and not col.startswith("an")]

    # Add only Q columns that are in your list and present in the table
    q_columns = [col for col in question_ids if col in all_columns]

    selected_columns = metadata_columns + q_columns

    # Build SELECT query
    column_str = ", ".join([f'"{col}"' for col in selected_columns])
    query = f"SELECT {column_str} FROM {table_name}"

    # Load and write to new table
    df_filtered = pd.read_sql(query, conn_src)
    df_filtered.to_sql(new_table_name, conn_dst, if_exists='replace', index=False)
    print(f"✔ Saved: {new_table_name} with {len(df_filtered.columns)} columns")

# Close connections
conn_src.close()
conn_dst.close()
```

```python
import sqlite3
import pandas as pd
from collections import defaultdict

# Function to rename case-insensitive duplicate columns
def rename_duplicate_columns_case_insensitive(df):
    seen = defaultdict(int)
    new_columns = []
    
    for col in df.columns:
        col_lower = col.lower()
        count = seen[col_lower]
        if count == 0:
            new_columns.append(col_lower)
        else:
            new_col = f"{col_lower}_dup{count}"
            print(f"📝 Renaming duplicate column: '{col}' → '{new_col}'")
            new_columns.append(new_col)
        seen[col_lower] += 1

    df.columns = new_columns
    return df

# Path to the cleaned database
cleaned_db = "/Users/duynguyen/Documents/Documents - DUY’s Mac mini/GitHub/ADS599_CapstoneProject/data/fevs_data_cleaned.db"
conn = sqlite3.connect(cleaned_db)

years = range(2018, 2025)
combined_df = pd.DataFrame()

for year in years:
    table_name = f"fevs_cleaned_{year}"
    try:
        df = pd.read_sql(f"SELECT * FROM {table_name}", conn)

        # Rename duplicate columns (case-insensitive) BEFORE adding 'year'
        df = rename_duplicate_columns_case_insensitive(df)

        # Add lowercase 'year' column only if it doesn't already exist
        if 'year' in df.columns:
            df.rename(columns={'year': 'year_dup_from_table'}, inplace=True)
            print(f"⚠️ 'year' column already existed in {table_name}, renamed to 'year_dup_from_table'")
        
        df['year'] = year  # Safe to add now

        combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"✔ Loaded: {table_name}, shape: {df.shape}")
    except Exception as e:
        print(f"❌ Failed to load {table_name}: {e}")

# Final check to ensure no duplicate column names before writing
combined_df = rename_duplicate_columns_case_insensitive(combined_df)

# Save to new consolidated table
combined_df.to_sql("fevs_cleaned_all_years", conn, if_exists="replace", index=False)
print(f"\n✅ Saved consolidated table 'fevs_cleaned_all_years' with shape: {combined_df.shape}")

conn.close()
```
