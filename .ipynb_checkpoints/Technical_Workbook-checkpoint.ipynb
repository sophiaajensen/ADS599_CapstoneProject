```python
#create a table to store data
conn = sqlite3.connect("/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS survey_question_master (
    ID INTEGER PRIMARY KEY,
    year INTEGER,
    question_ID INTEGER,
    section TEXT,
    question TEXT,
    note TEXT
);
""")

conn.commit()
conn.close()
```

```python
#Create a database of reference tables

# Path to save the database file
db_path = "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"

# Connect to SQLite database (will create if it doesn't exist)
conn = sqlite3.connect(db_path)

# Replace with your local CSV paths
excel_files = {
    "survey_2018": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2018 PRDF Codebook r2.xlsx",
    "survey_2019": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2019 PRDF Codebook.xlsx",
    "survey_2020": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2020 PRDF Codebook.xlsx",
    "survey_2021": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2021_OPM_FEVS_PRDF_Codebook_r2.xlsx",
    "survey_2022": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2022_OPM_FEVS_PRDF_Codebook_r2.xlsx",
    "survey_2023": "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/other material/OPM Docs/2023_OPM_FEVS_PRDF_Codebook.xlsx",
}

keyword = "survey"

# Loop and load only matching sheets
for table_name, file_path in excel_files.items():
    xl = pd.ExcelFile(file_path)
    
    # Find sheets with name 'survey'
    matching_sheets = [s for s in xl.sheet_names if keyword.lower() in s.lower()]
    
    if matching_sheets:
        sheet_to_use = matching_sheets[0]  
        df = xl.parse(sheet_name=sheet_to_use, header=0)
        df.to_sql(table_name, conn, if_exists='replace', index=False)
        print(f"Loaded '{table_name}' from sheet '{sheet_to_use}' into database.")
    else:
        print(f"No matching sheet found in {file_path}.")

conn.close()
```

```python
#clean up question bank for 2018 questions
db_path = "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Insert reference questions from 2018 survey
cursor.execute("""
    INSERT INTO survey_question_master (year, question_ID, section, question, note)
    SELECT 
        2018 AS year, 
        'Q' || "Item Number and Text" AS question_ID,
        CASE 
            WHEN "Item Number and Text" BETWEEN 1 AND 19 THEN 'My Work Experience'
            WHEN "Item Number and Text" BETWEEN 20 AND 29 THEN 'My Work Unit'
            WHEN "Item Number and Text" BETWEEN 30 AND 42 THEN 'My Agency'
            WHEN "Item Number and Text" BETWEEN 43 AND 52 THEN 'My Work Experience'
            WHEN "Item Number and Text" BETWEEN 53 AND 62 THEN 'Leadership'
            WHEN "Item Number and Text" BETWEEN 63 AND 71 THEN 'My Satisfaction'
            ELSE 'Unknown'
        END AS section,
        "Unnamed: 1" AS question,
        "CHANGES FROM 2017" AS note
    FROM survey_2018
    WHERE "Item Number and Text" IS NOT NULL
    AND "Item Number and Text" <> 'Item Number and Text'
""")

conn.commit()
conn.close()
```

```python
#clean up question bank for 2019 questions
db_path = "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Insert reference questions from 2019 survey
cursor.execute("""
    INSERT INTO survey_question_master (year, question_ID, section, question, note)
    SELECT 
        2019 AS year, 
        'Q' || "Item Number and Text" AS question_ID,
        CASE 
            WHEN "Item Number and Text" BETWEEN 1 AND 19 THEN 'My Work Experience'
            WHEN "Item Number and Text" BETWEEN 20 AND 28 THEN 'My Work Unit'
            WHEN "Item Number and Text" BETWEEN 29 AND 41 THEN 'My Agency'
            WHEN "Item Number and Text" BETWEEN 42 AND 52 THEN 'My Work Experience'
            WHEN "Item Number and Text" BETWEEN 53 AND 62 THEN 'Leadership'
            WHEN "Item Number and Text" BETWEEN 63 AND 71 THEN 'My Satisfaction'
            ELSE 'Unknown'
        END AS section,
        "Unnamed: 1" AS question,
        "CHANGES FROM 2018" AS note
    FROM survey_2019
    WHERE "Item Number and Text" IS NOT NULL
    AND "Item Number and Text" <> 'Item Number and Text'
""")

conn.commit()
conn.close()
```

```python
#clean up question bank for 2020 questions
db_path = "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Insert reference questions from 2020 survey
cursor.execute("""
    INSERT INTO survey_question_master (year, question_ID, section, question, note)
    SELECT 
        2020 AS year, 
        "VARIABLE" AS question_ID,
        CASE 
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 1 AND 8 THEN 'My Work Experience'
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 9 AND 13 THEN 'My Work Unit'
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 14 AND 18 THEN 'My Agency'
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 19 AND 25 THEN 'My Work Experience'
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 26 AND 32 THEN 'Leadership'
            WHEN CAST(SUBSTR("VARIABLE", 2) AS INTEGER) BETWEEN 33 AND 38 THEN 'My Satisfaction'
            ELSE 'COVID-19'
        END AS section,
        "Item Text" AS question,
        NULL AS note
    FROM survey_2020
    WHERE "ITEM TEXT" IS NOT NULL AND "Variable" NOT LIKE "D%"
""")

conn.commit()
conn.close()
```

```python
#clean up question bank for 2021 questions
db_path = "/Users/duynguyen/Documents/GitHub/ADS599_CapstoneProject/data/surveyquestions.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Insert reference questions from 2021 survey
cursor.execute("""
    INSERT INTO survey_question_master (year, question_ID, section, question, note)
    SELECT DISTINCT 
        2021 AS year, 
        "VARIABLE" AS question_ID,
        Null AS section,
        "Item Text" AS question,
        NULL AS note
    FROM survey_2021
    WHERE "ITEM TEXT" IS NOT NULL AND "Variable" NOT LIKE "D%"
""")

conn.commit()
conn.close()
```
