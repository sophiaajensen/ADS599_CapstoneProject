```python
# Categorize the survey data based on the codebook
categories = {
    "Demographics": ['DHISP', 'DDIS', 'DAGEGRP', 'DSEX', 'DSUPER', 'DFEDTEN', 'DMIL', 'DLEAVING'],
    "Golbal_Satisfaction": ['Q46', 'Q70', 'Q71', 'Q72'],
    "Performance_Confidence": ['Q20', 'Q21', 'Q22', 'Q23'],
    "Employee_Engagement": ['Q2', 'Q3', 'Q4', 'Q6', 'Q7', 'Q58', 'Q59', 'Q61', 'Q62', 'Q48', 'Q50', 'Q51', 'Q52', 'Q54'],
    "DEIA": ['Q73', 'Q74', 'Q75', 'Q76', 'Q77', 'Q78', 'Q79', 'Q80', 'Q81', 'Q82', 'Q83', 'Q84', 'Q85'],
    "Employee_Experience": ['Q86', 'Q87', 'Q88', 'Q89', 'Q90'],
    "Performance_Dimensions": {
        "Agile": ['Q13', 'Q14', 'Q27', 'Q29', 'Q30', 'Q64', 'Q28', 'Q31', 'Q42', 'Q65'],
        "Employee-Focused": ['Q1', 'Q10', 'Q25', 'Q66', 'Q67', 'Q36', 'Q43', 'Q44', 'Q5', 'Q34', 'Q49', 'Q63'],
        "Foundations": ['Q40', 'Q60', 'Q68', 'Q15', 'Q18', 'Q32', 'Q33', 'Q8', 'Q45', 'Q9', 'Q19', 'Q24'],
        "Goal_Oriented": ['Q11', 'Q53', 'Q26', 'Q38', 'Q12', 'Q55', 'Q56', 'Q17', 'Q35', 'Q69'],
        "Other": ['Q37', 'Q39', 'Q41', 'Q47']
    }
    
}

# Create new order 
new_order = (
    categories['Demographics'] +
    categories['Golbal_Satisfaction'] +
    categories['Performance_Confidence'] +
    categories['Employee_Engagement'] +
    categories['DEIA'] +
    categories['Employee_Experience'] +
    categories['Performance_Dimensions']['Agile'] +
    categories['Performance_Dimensions']['Employee-Focused'] +
    categories['Performance_Dimensions']['Foundations'] +
    categories['Performance_Dimensions']['Goal_Oriented'] +
    categories['Performance_Dimensions']['Other'] +
     ['POSTWT', 'Year']
)

df = df_2023[new_order]
```

```python
# Clean the data 
df = df.dropna(how='all')  # Drop rows where all elements are NaN
df = df.dropna(axis=1, how='all')  # Drop columns where all elements are NaN

# Map Likert scale values to numerical values
likert_map = {
    'A': 5,
    'B': 4,
    'C': 3,
    'D': 2,
    'E': 1,
    'X': None
}
# exclude last 2 columns
df = df.iloc[:, :-2].replace(likert_map).apply(pd.to_numeric, errors='coerce') # Exclude the last two columns (POSTWT, Year)

```
