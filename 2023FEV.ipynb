```python
# Categorize the survey data based on the codebook
categories = {
    "Demographics": ['DHISP', 'DDIS', 'DAGEGRP', 'DSEX', 'DSUPER', 'DFEDTEN', 'DMIL', 'DLEAVING'],
    "Golbal_Satisfaction": ['Q46', 'Q70', 'Q71', 'Q72'],
    "Performance_Confidence": ['Q20', 'Q21', 'Q22', 'Q23'],
    "Employee_Engagement": ['Q2', 'Q3', 'Q4', 'Q6', 'Q7', 'Q58', 'Q59', 'Q61', 'Q62', 'Q48', 'Q50', 'Q51', 'Q52', 'Q54'],
    "DEIA": ['Q73', 'Q74', 'Q75', 'Q76', 'Q77', 'Q78', 'Q79', 'Q80', 'Q81', 'Q82', 'Q83', 'Q84', 'Q85'],
    "Employee_Experience": ['Q86', 'Q87', 'Q88', 'Q89', 'Q90'],
    "Performance_Dimensions": {
        "Agile": ['Q13', 'Q14', 'Q27', 'Q29', 'Q30', 'Q64', 'Q28', 'Q31', 'Q42', 'Q65'],
        "Employee-Focused": ['Q1', 'Q10', 'Q25', 'Q66', 'Q67', 'Q36', 'Q43', 'Q44', 'Q5', 'Q34', 'Q49', 'Q63'],
        "Foundations": ['Q40', 'Q60', 'Q68', 'Q15', 'Q18', 'Q32', 'Q33', 'Q8', 'Q45', 'Q9', 'Q19', 'Q24'],
        "Goal_Oriented": ['Q11', 'Q53', 'Q26', 'Q38', 'Q12', 'Q55', 'Q56', 'Q17', 'Q35', 'Q69'],
        "Other": ['Q37', 'Q39', 'Q41', 'Q47']
    }
    
}

# Create new order 
new_order = (
    categories['Demographics'] +
    categories['Golbal_Satisfaction'] +
    categories['Performance_Confidence'] +
    categories['Employee_Engagement'] +
    categories['DEIA'] +
    categories['Employee_Experience'] +
    categories['Performance_Dimensions']['Agile'] +
    categories['Performance_Dimensions']['Employee-Focused'] +
    categories['Performance_Dimensions']['Foundations'] +
    categories['Performance_Dimensions']['Goal_Oriented'] +
    categories['Performance_Dimensions']['Other'] +
     ['POSTWT', 'Year']
)

df = df_2023[new_order]
```

```python
# Clean the data 
df = df.dropna(how='all')  # Drop rows where all elements are NaN
df = df.dropna(axis=1, how='all')  # Drop columns where all elements are NaN

# Map Likert scale values to numerical values
likert_map = {
    'A': 5,
    'B': 4,
    'C': 3,
    'D': 2,
    'E': 1,
    'X': None
}
# exclude last 2 columns
df = df.iloc[:, :-2].replace(likert_map).apply(pd.to_numeric, errors='coerce') # Exclude the last two columns (POSTWT, Year)

```

```python
# Drop all rows with NaN values
df_clean = df.dropna()

#df_clean.isna().sum()  # Verify that there are no NaN values left

df_clean.shape
```

Since the dataset has over 625,000 rows, removing rows with missing values is a simple strategy leaving us with 141,096 rows to analyze.



```python
# Rename demographic columns for clarity
df_clean.rename(columns={
    'DHISP': 'Hispanic_Origin',
    'DDIS': 'Disability_Status',
    'DAGEGRP': 'Age_Group',
    'DSEX': 'Gender',
    'DSUPER': 'Supervisory_Status',
    'DFEDTEN': 'Federal_Tenure',
    'DMIL': 'Military_Service_Status',
    'DLEAVING': 'Intent_to_Leave'
}, inplace=True)

df_clean.head()
```

```python
# Note: This function was obtained from Chat GPT and modified to fit the context of the FEVS data analysis

def survey_category_summary(df_clean, question_list, category_name):
    print(f"\n--- {category_name} ---")
    
    # Filter questions that exist in dataframe
    existing_questions = [q for q in question_list if q in df_clean.columns]
    if not existing_questions:
        print("No questions found in data for this category.")
        return
    
    # Convert to numeric and handle errors if any
    df_numeric = df_clean[existing_questions].apply(pd.to_numeric, errors='coerce')
    
    print(df_numeric.describe())
    
    # Calculate mean
    mean_scores = df_numeric.mean().sort_values(ascending=False)
    
    # Plotting the mean scores
    plt.figure(figsize=(10, 4))
    sns.barplot(x=mean_scores.index, y=mean_scores.values, palette="viridis")
    plt.title(f"Mean Scores: {category_name.replace('_', ' ')}")
    plt.ylabel("Average Response (Likert Scale)")
    plt.xlabel("Survey Questions")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Generate summary for each category in the survey
for category, questions in categories.items():
    if isinstance(questions, list):
        survey_category_summary(df_clean, questions, category)
```

Overall, it appears that for each category the average is over 3 and 4 indicating that the employees are having more positive experience.
